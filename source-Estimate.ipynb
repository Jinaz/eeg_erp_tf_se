```python
import os.path as op
import mne
from mne.datasets import fetch_fsaverage
import readData as rd
from matplotlib import pyplot as plt 
```

```python
from surfer import Brain
```

```python
fsdir = fetch_fsaverage()
subjectsdir = op.dirname(fsdir)
subjectid = '001'

subject = 'fsaverage'
trans = 'fsaverage'
src = op.join(fsdir, 'bem', 'fsaverage-ico-5-src.fif')
bem = op.join(fsdir, 'bem', 'fsaverage-5120-5120-5120-bem-sol.fif')
```

```python
raw, evts, evtsdict = rd.readBids(subjectid)
epochs = rd.get_epoched_Data(raw, evts, evtsdict)
conditions = ['faces', 'cars', 'scrambled_faces', 'scrambled_cars']

epochs_faces = epochs.copy()[conditions[0]]
epochs_cars = epochs.copy()['cars']
epochs_cars.set_eeg_reference(projection=True)
epochs_faces.set_eeg_reference(projection=True)



evfaces = epochs_faces.average()
evcars = epochs_cars.average()

comparison = mne.combine_evoked([evfaces,evcars], [1,-1])
inittime = comparison.copy().crop(0.11,0.18).pick(['PO8']).get_peak()[1]
```

```python
fwd = mne.make_forward_solution(epochs.info, trans=trans, src=src, bem=bem, eeg=True, mindist=5.0)
```

```python
noise_cov = mne.compute_covariance([epochs_faces, epochs_cars], tmax=0)
noise_cov.plot(epochs.info)
```

```python
from mne.minimum_norm import make_inverse_operator, apply_inverse

inv_default = make_inverse_operator(epochs.info, fwd, noise_cov, loose=0.2, depth=0.8)
inv_loose0 = make_inverse_operator(epochs.info, fwd, noise_cov, loose=0, depth=0.8)
inv_loose1 = make_inverse_operator(epochs.info, fwd, noise_cov, loose=1, depth=0.8)
inv_depth0 = make_inverse_operator(epochs.info, fwd, noise_cov, loose=0.2, depth=0.)
```

```python
snr = 10.0
lambda2 = 1.0 / snr ** 2
stc = apply_inverse(comparison, inv_default, lambda2, 'MNE')#, pick_ori='vector')
```

# check if peaks are at place


```python
brain = stc.plot(time_viewer=True,hemi="both")
```

```python
brain_kwargs = dict( show=False)

# plot it
#h = stc.plot(**brain_kwargs)

h = mne.viz.plot_source_estimates(stc, initial_time=inittime, time_viewer=False,time_unit='s',views="lateral",hemi='split', size=(800, 400), show_traces=False, brain_kwargs=brain_kwargs)
img = stc_plot2img(h,closeAfterwards=True)
plt.imshow(img)
plt.axis('off')
plt.savefig("img/sub-{}/sub-{}_SourceEstimateAt-{}.png".format(subjectid, subjectid, inittime))
```

Taking the plotting tooling to make some compact images

```python
def stc_plot2img(h,title="SourceEstimate",closeAfterwards=False,crop=True):
    h.add_text(0.1, 0.9, title, 'title', font_size=16)
    screenshot = h.screenshot()
    if closeAfterwards:
        h.close()

    if crop:
        nonwhite_pix = (screenshot != 255).any(-1)
        nonwhite_row = nonwhite_pix.any(1)
        nonwhite_col = nonwhite_pix.any(0)
        screenshot = screenshot[nonwhite_row][:, nonwhite_col]
    return screenshot
```

# Using the methods from exercise to get some compact images

```python
epochs_faces.load_data()
epochs_cars.load_data()
#brain_kwargs = dict(views='flat',surface="flat", hemi='split', size=(800, 400), initial_time=inittime,time_viewer=False,time_unit='s',show_traces=False)
# plotting parameters
brain_kwargs = dict(views="lateral",hemi='split', size=(800, 400), initial_time=inittime,time_viewer=False,time_unit='s',show_traces=False)

# getting larger figures
plt.rcParams['figure.dpi'] = 100 
plt.rcParams["figure.figsize"] = (20,15)

# a list of all our inverse operators
sourceList = [inv_default,inv_loose0,inv_loose1,inv_depth0]

#loop over inverse operators
for (ix,inv) in enumerate(sourceList):

    # calculate the stc
    s = apply_inverse(comparison, inv, lambda2, 'MNE', verbose=True)#, pick_ori='vector')
    
    # plot it
    h = s.plot(**brain_kwargs)

    # grab a png from it
    img = stc_plot2img(h,closeAfterwards=False)

    # add the png to a subplot
    ax = plt.subplot(2,2,ix+1).imshow(img)
    plt.axis('off')
    plt.savefig("img/sub-{}/sub-{}_sourceEstimate.png".format(subjectid,subjectid))
```
